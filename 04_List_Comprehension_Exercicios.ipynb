Criando um dicionário de dicionários contendo a contagem de letras em cada nome.

```python
def contador(lista):
    lista_string = []
    contador_nome ={}

    for nome in lista:
        string_lista_letras = list(nome.lower())
              
        elementos_distintos = []
        contagem_elementos = []

        for n in string_lista_letras:
            if n in string_lista_letras:
                if n in elementos_distintos:
                    pass
                else:
                    elementos_distintos.append(n)

        for n in elementos_distintos:
            contagem = 0
            for elemento in string_lista_letras:
                if elemento == n:
                    contagem += 1
            contagem_elementos.append(contagem)
        
        dict_contagem = dict(zip(elementos_distintos, contagem_elementos))

        contador_nome[nome] = dict_contagem      

    return contador_nome

contador(nomes)
```
