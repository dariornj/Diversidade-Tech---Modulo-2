## List Comprehension - Desafio

**Enunciado**

Crie um sistema de cadastro de produtos. Neste sistema podemos:

Adicionar um novo produto - Remover um produto da base - Consultar quais são os produtos cadastrados - Consultas quais os produtos cadastrados e suas quantidades disponíveis -  Adicionar informações extras por produto (descrição por exemplo) - Adicionar ao estoque de um produto - Remover do estoque um produto (nota, o total em estoque não pode ser menor que 0)

Para tal crie as seguintes funções:

cadastre_produto - delete_produto - adicione_produto_estoque - remova_produto_estoque - consulte_produtos - consulte_quantidade - consulte_descricao_produto - ative_sistema

Essa função irá gerenciar todas as funções acima (como um sistema central)

Os atributos possíveis são:

Nome do produto - Quantidade do produto - descrição - Informações adicionais


```python
def cadastre_produto():
  """Essa função cadastra um novo produto com os campos:
    - nome do produto (obrigatório)
    - quantidade (opcional)
    - descrição (opcional)
    - outros campos
  """
  continuar_cadastro = True
  banco_dados= {}

  while continuar_cadastro:
    produto = input('Digite o nome do produto: ')
    if produto == '':
      print('Nome do produto é obrigatório.')
      break
    quantidade = int(input(f'Digite a quantidade de {produto}: '))
    descricao = input(f'Digite a descricao de {produto}')
    info_add = input(f'Digite informações adicionais sobre {produto}')
    banco_dados[produto] = {'Quantidade': quantidade, 'Descrição': descricao, 'Informações adicionais': info_add}

    continua = input('Deseja cadastrar outro produto? (1) Sim (2) Não')
    if continua == '2':
        continuar_cadastro = False
        return banco_dados
```

```python
## Para testes
banco_dados = {'caixa': {'Quantidade': 20,
  'Descrição': 'util',
  'Informações adicionais': ''},
 'estojo': {'Quantidade': 30, 'Descrição': '', 'Informações adicionais': ''},
 'lapis': {'Quantidade': 10, 'Descrição': '', 'Informações adicionais': ''}}
```

```python
def delete_produto(banco_dados):
  """ Essa função deleta um produto da base pelo `nome do produto`
  """
  
  produto_deletado = input('Qual produto deseja remover? ')
  {banco_dados.pop(produto) for produto in banco_dados if produto.lower() == produto_deletado.lower()}

  return banco_dados

novo_banco_dados = delete_produto(banco_dados)
```

```python
def adicione_produto_estoque(banco_dados):
  """ Essa função adiciona ao estoque uma quantidade de um dado produto
      Nota: Não pode ser aceito quantidade negativas
  """

  produto_selecionado = input('Qual produto deseja adicionar estoque? ')
  quantidade_add = int(input('Digite a quantidade a ser adicionada: '))

  for produto in banco_dados:
    if quantidade_add < 0:
        print('Quantidade adicionada deve ser positiva.')
        break

    if produto.lower() == produto_selecionado.lower():
        banco_dados[produto]['Quantidade'] = banco_dados[produto]['Quantidade'] + quantidade_add
  
  return banco_dados

```

```python
def remova_produto_estoque(banco_dados):
  """ Essa função remove do estoque uma quantidade de um dado produto
      Nota: Não pode ser aceito quantidade negativas
  """

  produto_selecionado = input('Qual produto deseja remover estoque? ')
  quantidade_removida = int(input('Digite a quantidade a ser removida: '))

  for produto in banco_dados:
    if quantidade_removida < 0:
        print('Quantidade removida deve ser positiva.')
        break

    if produto.lower() == produto_selecionado.lower():
        qtd_inicial = banco_dados[produto]['Quantidade']
        qtd_final = qtd_inicial - quantidade_removida
        if qtd_final >= 0:
          banco_dados[produto]['Quantidade'] = qtd_inicial - quantidade_removida
        else:
          print(f'Quantidade removida maior que estoque de {qtd_inicial} unidades.')
  
  return banco_dados  
```

```python
def consulte_produtos(banco_dados):
  """ Essa função mostra os produtos disponíveis no sistema (somente nome)
  """
  {print(produto) for produto in banco_dados}
```

```python
def consulte_quantidade(banco_dados):
  """ Essa função mostra os produtos e a quantidade disponíveis no sistema
  """
  {print(f"{produto}: {banco_dados[produto]['Quantidade']} unidades no estoque.") for produto in banco_dados}
```

```python
def consulte_descricao_produto(banco_dados):
  """ Essa função mostra a descrição e as Informações adicionais de um dado produto
  """
  produto_selecionado = input('Qual produto deseja ver a descrição? ')
  if produto_selecionado in list(banco_dados.keys()):
    {print(f"{produto}: {banco_dados[produto]['Descrição']}") for produto in banco_dados if produto.lower() == produto_selecionado.lower()}
  else:
    print('Produto não encontrado no banco de dados. ')
```

```python
def ative_sistema():
  """ Essa função aceita as interações do usuário, coordenando qual ação deve ser tomada
      Cada ação refere-se as funções desenvolvidas acima.
      Nota: o que fazer se for inserida uma ação inválida?
  """

continua = True
while continua:
  action = int(input('O que deseja fazer? \n(1) Cadastrar produto \n(2) Deletar produto\n (3) Adicionar produto no estoque \n(4) Remover produto do estoque\n (5) Consultar produtos\n (6) Consultar quantidade de produtos\n (7) Consultar descrição de um produto\n (8) Encerrar programa.'))

  if action == 1:
    banco_dados = cadastre_produto()
  elif action == 2:
    banco_dados = delete_produto(banco_dados)
  elif action == 3:
    banco_dados = adicione_produto_estoque(banco_dados)
  elif action == 4:
    banco_dados = remova_produto_estoque(banco_dados)
  elif action == 5:
    consulte_produtos(banco_dados)
  elif action == 6:
    consulte_quantidade(banco_dados)
  elif action == 7:
    consulte_descricao_produto(banco_dados)
  elif action == 8:
    print('Obrigado pelo uso. ')
    continua = False
  else:
    print('Opção inválida, escolha novamente.')
  
```
