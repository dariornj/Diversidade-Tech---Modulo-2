**Enunciado**

Modifique a função cadastrar_usuario abaixo em que serão coletados as seguintes informações a partir da entrada da pessoa usuária (input):
- Nome
- Idade
- Sexo


Nesse cadastro busque utilizar namedtuple ou tuplas. Além de aceitar múltiplas entradas do usuário (aceite a opção cadastrar, parar, e lide com uma opção invalida)

Agora crie uma função que permita descobrir a idade média de pessoas cadastradas pelo sexo (para manter simples, masculino e feminino, representado por m e f, respectivamente).

Nessa função, perceba que o sexo pode ter sido cadastrado de forma incorreta, como proceder?

Na função calcule_media_idade_por_sexo o retorno deve ser uma tupla de tuplas no formato

(('m', <media_masculino>), ('f', <media_feminino>))


```python
from collections import namedtuple

def cadastrar_usuario():
  continuar = True

  cadastros = []
  Cliente = namedtuple('Cliente', ['nome', 'idade', 'sexo'])
  
  while continuar:
    cadastrar = input('Deseja cadastrar novo usuário?')
    if cadastrar.lower() in ['sim', 's', 'yes', 'y']:
      nome = input('Digite seu nome: ')
      idade = int(input('Digite sua idade: '))
      sexo = input('Digite o sexo (m/f)')
      cadastro = Cliente(nome=nome, idade=idade, sexo=sexo)
      cadastros.append(cadastro)
    elif cadastrar.lower() in ['nao', 'n', 'não', 'no']:
      print(f'Cadastro finalizado, inserido {len(cadastros)} novos')
      continuar = False
    else: 
      print(f'Opção invalida. Foi inserido "{cadastrar}",\
digite [sim, s, yes, y] para cadastrar um cliente ou [não, nao, n, no] para parar o cadastro')

  return cadastros
```

```python
cadastros = cadastrar_usuario()
```

```python
def calcule_media_idade_por_sexo(cadastros):
    idade_feminino = []
    idade_masculino = []
    incorreto = []

    for cliente in cadastros:
        idade = cliente.idade
        if cliente.sexo.lower() == 'm':
            idade_masculino.append(idade)
        elif cliente.sexo.lower() == 'f':
            idade_feminino.append(idade)
        else:
            incorreto.append(idade)

    if len(idade_masculino) > 0 :
        media_masculino = sum(idade_masculino)/len(idade_masculino)
    else:
        media_masculino = 'Não existe'

    if len(idade_feminino) > 0:
        media_feminino = sum(idade_feminino)/len(idade_feminino)
        
        n_incorretos = len(incorreto)

    return (('f', media_feminino), ('m', media_masculino), ('numero de entradas incorretas:', n_incorretos))
```
